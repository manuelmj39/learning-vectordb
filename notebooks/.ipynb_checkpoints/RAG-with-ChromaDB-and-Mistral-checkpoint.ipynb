# Retrieval Augmented Generation 
Retrieval augmented generation is a technique of enriching your language model outputs by retrieving contextual information from an external data source, and including that information as part of your language model prompts. The idea is that when you augment a language model prompt with meaningful external data, the language model is able to respond with deeper understanding and relevance.

**In this notebook we will:**
1. Construct a dataset that can be indexed and queried.
2. Load the dataset into ChromaDB (a vector store).
3. Run some test queries against ChromaDB and visualize what is in the database.
4. Bring it all together. Query ChromaDB for 10 related popular titles, then prompt mistral-7b-instruct to suggest new titles, inspired by the related popular titles.

We will be using a car reviews dataset for the tutorial. Link: https://www.kaggle.com/datasets/ankkur13/edmundsconsumer-car-ratings-and-reviews

## Read Data

```python
import pandas as pd

file_path = '/Users/manueljohn/Training/NLP/learning-vectordb/data/car-review/Scraped_Car_Review_hyundai.csv'

#read file
hyundai_review_df = pd.read_csv(file_path, 
                                lineterminator='\n')

req_cols = ['Review_Date', 'Author_Name', 'Vehicle_Title', 'Review_Title', "Review", "Rating\r"]
hyundai_review_df = hyundai_review_df[req_cols]

#change column name
hyundai_review_df.rename(columns={"Rating\r": 'Rating'}, inplace=True)

print("Shape of Dataframe: ", hyundai_review_df.shape)

hyundai_review_df.head()
```

### Missing Values

```python
#Column-wise Missing values
missing_values = hyundai_review_df.isnull()

print("Missing Values:\n\n", missing_values.sum(), '\n')
```

**No Null values exists**

 ### Check Duplicates

```python
print("Duplicated: ", hyundai_review_df.duplicated().sum())

#remove complete dupes
hyundai_review_df = hyundai_review_df.drop_duplicates(keep='first')\
                    .reset_index(drop=True)

hyundai_review_df
```

```python
cols = ['Review_Date', 'Author_Name', 'Vehicle_Title']

#Mutiple entries by same person
hyundai_review_df[hyundai_review_df.duplicated(subset=cols, keep=False)]

#keep last
hyundai_review_df = hyundai_review_df.drop_duplicates(keep='last', subset=cols)\
                    .reset_index(drop=True)

hyundai_review_df
```

### Stats and Viz

```python
#Hyundai Total Review
hyundai_review_df['Rating'].describe()
```

```python
import statistics, matplotlib.pyplot as plt  

#mean of car models
full_dict = dict(hyundai_review_df['Vehicle_Title'].value_counts())

plt.title("Car Sold Counts")
plt.hist(full_dict.values(), bins=5, density=True)

plt.show()
```

```python
#Sorting Bad Review Models
lt_avg = hyundai_review_df[hyundai_review_df['Rating']<3.0]

lt_dict = dict(lt_avg['Vehicle_Title'].value_counts())

#vehicles w/t more than 30% poor review
poor_revi_dict = {i: str((lt_dict[i]/full_dict[i], lt_dict[i], full_dict[i]))
                  for i in lt_dict 
                  if (lt_dict[i]/full_dict[i] > 0.3 and full_dict[i]>10)} 

poor_revi_dict
```

Seems like most bad reviewed models are **Tucson and Sonata**
